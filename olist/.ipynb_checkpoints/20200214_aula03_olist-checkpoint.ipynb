```python
def bar_plot(x, y, df, ax, colors='Blues_d', hue=False, value=False):
    """
    This function plots, and customize a bar chart

    Input:
        x: feature to be plotted on x axis
        y: feature to be plotted on y axis
        df: DataFrame object with features used on x and y
        colors: standard palette is "Blues_d"
        hue: separation value, standard is False
        value: flag for defining if the data labels on bars will be the value (True) or percentual (False)
        
    Returns:
        None
    """
    # Counting numerical feature (x or y)
    try:
        ncount = sum(df[y])
    except:
        ncount = sum(df[x])
    
    # Plotting
    if hue != False:
        ax = sns.barplot(x=x, y=y, data=df, palette=colors, hue=hue, ax=ax, ci=None)
    else:
        ax = sns.barplot(x=x, y=y, data=df, palette=colors, ax=ax, ci=None)

    # Customizing data labels (values of percents)
    for p in ax.patches:
        xp=p.get_bbox().get_points()[:,0]
        yp=p.get_bbox().get_points()[1,1]
        if value:
            ax.annotate('{:.2f}k'.format(yp/1000), (xp.mean(), yp), 
                    ha='center', va='bottom')
        else:
            ax.annotate('{:.1f}%'.format(100.*yp/ncount), (xp.mean(), yp), 
                    ha='center', va='bottom')
            
def donut_plot(col, ax, df, labels, text='', flag_ruido = 0,
               colors=['navy', 'lightsteelblue', 'lightgreen', 'crimson', '']):
    """
    This function plots a customized donut plot
    
    Input:
        col: coluna a ser analisada e plotada no gráfico de rosca
        ax: matplotlib axis
        df: DataFrame with data
        labels: list of labels to be plotted
        text: text to be plotted on the center of the donut
        flag_ruido: thie parameter indicates the number of labels to be filtered from plot
        colors: list of colors (4 colors from default)
    """
    flag_ruido = flag_ruido * -1
    if flag_ruido < 0:
        sizes = df[col].value_counts().values[:flag_ruido]
        labels = labels[:flag_ruido]
    else:
        sizes = df[col].value_counts().values
    center_circle = plt.Circle((0,0), 0.80, color='white')
    ax.pie(sizes, labels=labels, colors=colors, autopct='%1.2f%%')
    ax.add_artist(center_circle)
    kwargs = dict(size=20, fontweight='bold', va='center')
    ax.text(0, 0, text, ha='center', **kwargs)

def format_spines(ax, right_border=True):
    """
    This function is responsible for format axis from graphs
    
    Input:
        ax: matplotlib axis
        right_border: boolean flag for plot the right border of graphs
    
    Output:
        None
    """    
    ax.spines['bottom'].set_color('#CCCCCC')
    ax.spines['left'].set_color('#CCCCCC')
    ax.spines['top'].set_visible(False)
    if right_border:
        ax.spines['right'].set_color('#CCCCCC')
    else:
        ax.spines['right'].set_color('#FFFFFF')
    ax.patch.set_facecolor('#FFFFFF')
```

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import datetime as dt
%matplotlib inline
plt.rcParams["figure.figsize"] = (20,10)
```

### importação dos arquivos

```python
products = pd.read_csv('olist_products_dataset.csv')
products.name = 'products'

orders = pd.read_csv('olist_orders_dataset.csv')
orders.name = 'orders'

order_items = pd.read_csv('olist_order_items_dataset.csv')
order_items.name = 'order items'

sellers = pd.read_csv('olist_sellers_dataset.csv')
sellers.name = 'sellers'

customers = pd.read_csv('olist_customers_dataset.csv')
customers.name = 'customers'

order_payments = pd.read_csv('olist_order_payments_dataset.csv')
order_payments.name = 'order payments'

order_reviews = pd.read_csv('olist_order_reviews_dataset.csv')
order_reviews.name = 'order reviews'

geo = pd.read_csv('olist_geolocation_dataset.csv')
geo.name = 'geolocation'
```

### resumindo um dataframe

```python
def exploreFrequencies(df):
    print("{0:30} {1:25} {2:25}".format(df.name, "unique values", "missing"))
    for i in df:
        print("{0:30} {1:25} {2:25}".format(i, df[i].nunique(), df[i].isna().sum()))
```

```python
tabelao = orders.merge(order_items, on = 'order_id', how = 'left')
tabelao = tabelao.merge(order_payments, on = 'order_id', how = 'outer', validate = 'm:m')
tabelao = tabelao.merge(order_reviews, on = 'order_id', how = 'outer')
tabelao = tabelao.merge(products, on = 'product_id', how = 'outer')
tabelao = tabelao.merge(customers, on = 'customer_id', how = 'outer')
tabelao = tabelao.merge(sellers, on = 'seller_id', how = 'outer')
#tabelao = tabelao.merge(geo, left_on = 'customer_zip_code_prefix', right_on = 'geolocation_zip_code_prefix', how = 'left')
```

```python
payments_labels = tabelao['payment_type'].value_counts().index
colors = ['navy', 'steelblue', 'lightsteelblue', 'powderblue']
donut_plot('payment_type', plt.gca(), 
           tabelao,
           labels=payments_labels,
           colors=colors,
           text='payment type',
           flag_ruido=1) #flag_ruido -> quantas categorias das menos representativas são retiradas
```

```python
time_cols = ['order_purchase_timestamp', 'order_approved_at', 'order_delivered_carrier_date',
           'order_delivered_customer_date', 'order_estimated_delivery_date']
for col in time_cols:
    tabelao[col] = pd.to_datetime(tabelao[col],  errors='coerce' )

tabelao['order_purchase_year'] = \
tabelao['order_purchase_timestamp'].apply(lambda x: x.year)

# Creating new month column
tabelao['order_purchase_month'] = \
tabelao['order_purchase_timestamp'].apply(lambda x: x.month)

# Creating new day of week column
tabelao['order_purchase_dayofweek'] = \
tabelao['order_purchase_timestamp'].apply(lambda x: x.dayofweek)

tabelao['order_purchase_month'] = tabelao['order_purchase_month'].astype(str).\
apply(lambda x: '0' + x if len(x) == 1 else x)
```

```python
texto = str()

for x in bad_review['review_comment_message']:
    texto = texto + str(x)
```

```python
texto
```

```python
txt = texto.lower().replace('\n',' ') \
                   .replace(',',"") \
                   .replace('.','') \
                   .replace('?','') \
                   .replace(":",'')\
                   .replace("!",'')
                
output_str  = str()  
output  = []

for p in txt.split():
    if not (p in palavra_parada or p[0:4] == 'http'):
        output_str = output_str + ' ' + p
        output.append(p)
```

```python
import re
from collections import Counter
import matplotlib.pyplot as plt

def words(text): return re.findall(r'\w+', text.lower())

WORDS = Counter(words(output_str))

WORDS.most_common
```

```python
wordcloud = wc.WordCloud(max_font_size=75, max_words=100, background_color="white").generate(" ".join(output))

plt.figure()
plt.imshow(wordcloud, interpolation="bilinear")
plt.axis("off")
plt.show()
```
